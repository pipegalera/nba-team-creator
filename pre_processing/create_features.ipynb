# Pre-processing and cleaning data

```python
# Remove combination of stats of different teams
adv_stats = adv_stats[adv_stats.Tm != "TOT"]
# Create a unit indicator
adv_stats = create_unit_indicator(adv_stats)
# Create summations of PER by players units and their rank
teams_per = create_summed_PER_by_units(adv_stats)
# Merge teams_per and team data
teams_data = teams_data.merge(teams_per)
```
